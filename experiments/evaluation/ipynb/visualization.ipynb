```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# plot omega

```python
participent_nr = 5  # choose participent

bag_name = "sloppy_joystick"

twist_topic_list = ['sloppy_joy_cmd', 'nav_cmd', 'joy_cmd', 'homing_cmd', 'cmd_vel_rc100', 'cmd_vel', 'col_cmd']

plt.figure(figsize=(11.69,8.27))
for topic_name in twist_topic_list:
    twist_df = pd.read_csv("../df/participent_" + str(participent_nr) + "/" + bag_name + "/twist_" + topic_name + ".csv")
    plt.plot(twist_df['time'], twist_df['omega'], label=topic_name)
plt.legend(loc="lower left")
plt.show()
```

# plot odom

```python
participent_numbers = [1,2,3,4,5] 

plt.figure(figsize=(11.69,8.27))
for participent_nr in participent_numbers:
    odom_df = pd.read_csv("../df/participent_" + str(participent_nr) + "/" + bag_name + "/odom.csv")
    plt.plot(odom_df['x'], odom_df['y'], label="participent "+str(participent_nr))
plt.legend(loc="lower left")
plt.axis('equal')
plt.show()
```
